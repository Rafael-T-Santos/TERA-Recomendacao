```python
columns = ['ID', 'PREGO', 'MARTELO', 'TINTA', 'PINCEL', 'ESPATULA', 'BRINDE']
dataset = [[1, 1, 1, 1, 1, 0, 1],
           [2, 1, 1, 1, 0, 0, 1],
           [3, 1, 1, 0, 0, 0, 1],
           [4, 1, 0, 0, 0, 0, 1],
           [5, 0, 1, 1, 1, 1, 1],
           [6, 0, 0, 1, 1, 1, 1],
           [7, 0, 0, 0, 1, 1, 1],
           [8, 0, 0, 0, 0, 1, 1]]

df2 = pd.DataFrame(dataset, columns=columns)
if 'ID' in df2.columns: del df2['ID'] # ID não é relevante para o Apriori
```

```python
df2
```

```python
frequent_itemsets = fpgrowth(df2, min_support=0.25, use_colnames=True)
frequent_itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
rules['length'] = rules['antecedents'].apply(lambda x: len(x))
rules['length2'] = rules['consequents'].apply(lambda x: len(x))
rules[(rules['length'] == 1) & (rules['length2'] == 1)]
```
